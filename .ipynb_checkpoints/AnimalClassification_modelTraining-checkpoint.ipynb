# Animal Classification using with Tensorflow

References: https://www.kaggle.com/code/nilawilson/animal-species-image-classification-dnn

Dataset: https://www.kaggle.com/datasets/alessiocorrado99/animals10/data

### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

### Data Preprocessing

```python
import os

dataset_path = 'animals_dataset'
print(os.listdir(dataset_path)) 
```

```python
class_names = ['dog', 'horse', 'elephant', 'butterfly', 'chicken', 'cat', 'cow', 'sheep', 'spider', 'squirrel']
```

```python
# Data augmentation
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1.0/255, zoom_range=0.2, horizontal_flip=True, validation_split=0.2)

# classifying the dataset into train and test data.
train_generator = train_datagen.flow_from_directory(dataset_path, target_size=(224, 224), class_mode='categorical', subset='training')
validation_generator = train_datagen.flow_from_directory(dataset_path, target_size=(224, 224), class_mode='categorical', subset='validation')
```

### Model Building

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.models import Model
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Conv2D, GlobalAveragePooling2D, Flatten, Dense, Dropout, Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# Input layer
input_tensor = Input(shape=(224, 224, 3))

# Loading the MobileNetV2 model, excluding the top layer
base_model = MobileNetV2(weights='imagenet', include_top=False, input_tensor=input_tensor)

# Freezing the base model layers
for layer in base_model.layers:
    layer.trainable = False

# Adding custom top layers
x = GlobalAveragePooling2D()(base_model.output)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
output_tensor = Dense(10, activation='softmax')(x)  

# Creating the model
model = Model(inputs=input_tensor, outputs=output_tensor)

model.summary()
```

```python
# Compiling the model
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Callbacks for early stopping 
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# Training the model
history = model.fit(train_generator, validation_data=validation_generator, epochs=6, callbacks=[early_stopping])
```

```python
# Evaluating the model on validation data
val_loss, val_accuracy = model.evaluate(validation_generator)
val_accuracy
```

```python
# Plotting loss and accuracy of validation and test data
h = history.history
plt.plot(h['loss'], c='red', label='Training Loss')
plt.plot(h['val_loss'], c='red', linestyle='--', label='Validation Loss')
plt.plot(h['accuracy'], c='blue', label='Training Accuracy')
plt.plot(h['val_accuracy'], c='blue', linestyle='--', label='Validation Accuracy')
plt.xlabel("Number of Epochs")
plt.legend()
```

### Predictions

```python
from tensorflow.keras.preprocessing import image

def load_and_preprocess_image(img_path, target_size=(224, 224)):
    img = image.load_img(img_path, target_size=target_size)
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array /= 255.0  
    return img_array

def predict_image(img_path):
    img_array = load_and_preprocess_image(img_path)
    prediction = model.predict(img_array)
    
    img = image.load_img(img_path)
    plt.imshow(img)
    
    plt.title(f"Prediction: {class_names[np.argmax(prediction)]}")
    plt.show()
```

```python
predict_image('squirrel.jpg')
```
